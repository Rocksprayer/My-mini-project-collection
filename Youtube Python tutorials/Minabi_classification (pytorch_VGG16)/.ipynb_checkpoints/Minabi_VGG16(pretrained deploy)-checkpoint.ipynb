```python
import numpy as np

import torch
import torchvision
from torchvision import models, transforms

import PIL 
from PIL import Image

import matplotlib.pyplot as plt

import json
```

```python
print("Ptorch Version", torch.__version__)
print("Ptorchvision Version", torchvision.__version__)
```

```python
use_pretrain = True
net = models.vgg16(pretrained = use_pretrain)
net.eval()
print(net)
```

```python
class BaseTransform():
    def __init__(self,resize,mean,std):
        self.base_transform = transforms.Compose([
            transforms.Resize(resize),
            transforms.CenterCrop(resize),
            transforms.ToTensor(),
            transforms.Normalize(mean,std)
        ])
    def __call__ (self,img):
        return self.base_transform(img)
```

```python
image_file_path =".\data\husky.jpg"
img = Image.open(image_file_path)

plt.imshow(img)
plt.show()

resize = 224
mean=(0.485, 0.456, 0.406)
std=(0.229, 0.224, 0.225)


transforming  = BaseTransform(resize , mean,std)
img_transform = transforming(img)

img_transform = img_transform.numpy().transpose(1,2,0)
img_transform = img_transform.clip(0,1)
plt.imshow(img_transform)
plt.show()
```

```python
class Predictor():
    def __init__(self,class_index):
         self.class_index = class_index
    def predict_max(self,net_out):
        maxid = np.argmax(net_out.detach().numpy())
        predicted_label_name = self.class_index[str(maxid)]
        return predicted_label_name
```

```python
class_index =json.load(open("./imagenet_class_index.json"))
```

```python
predictor = Predictor(class_index)
```

```python
image_file_path =".\data\husky.jpg"
img = Image.open(image_file_path)

resize = 224
mean=(0.485, 0.456, 0.406)
std=(0.229, 0.224, 0.225)


transforming  = BaseTransform(resize , mean,std)
img_transform = transforming(img)
img_transform = img_transform.unsqueeze(0)
net_out=net(img_transform)

result = predictor.predict_max(net_out=net_out)
```

```python
print("result is {}".format(result[1]))
```
